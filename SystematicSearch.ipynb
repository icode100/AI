This Module shall consist of Systematic search algorithms namely  A*, IDA*, RBFS. We shall solve 8 puzzle problem for this module

```python
# defining the state -> actions, cost and parent the gx hx and fx values and the expansion function
from copy import deepcopy
from heapq import heappop, heappush
from typing import Callable
class State:
    def __init__(self, state:list[list[int]],heuristicComputer,goal:list[list[int]], parent:'State'= None, gx:int = 0, hx:int = 0):
        self.state = state
        self.parent = parent
        self.gx = gx
        self.hx = hx
        self.isGoal = True if self.state==goal else False
        self.fx = self.hx+self.gx
        self.goal = goal
        self.actions = [(-1,0),(1,0),(0,-1),(0,1)]
        self.blank = (-1,-1)
        self.heuristic = heuristicComputer
        for i in range(3):
            for j in range(3):
                if self.state[i][j] == 0:
                    self.bank = (i,j)
                    break
        
        flat_state = [n for row in self.state for n in row]
        inversions = 0
        for i in range(8):
            for j in range(i + 1, 9):
                if flat_state[i] and flat_state[j] and flat_state[i] > flat_state[j]:
                    inversions += 1
        self.solvability = inversions % 2 == 0
        def __le__(self,other):
            return self.fx<=other.fx
        

    def expand(self)->list['State']:
        result = list()
        for action in self.actions:
            nr = self.blank[0]+action[0]
            nc = self.blank[1]+action[1]
            if 0<=nr and nr<3 and 0<=nc and nc<3:
                newState = deepcopy(self.state)
                newState[nr][nc],newState[self.blank[0]][self.blank[1]] = newState[self.blank[0]][self.blank[1]],newState[nr][nc]
                result.append(State(newState,self.heuristic,self.goal,self,self.gx+1,self.heuristic(self.state,self.goal))) 
        return result


class Algorithms:
    def misplacedTiles(self,matrix,goal):
        result:int = 0
        for i in range(3):
            for j in range(3):
                if matrix[i][j]!=goal[i][j]:
                    result+=1
        return result
    def __init__(self,start,goal):
        self.s = State(start,self.misplacedTiles,goal)
    def AStar(self):
        pq:list[State] = list()
        heappush(pq,self.s)
        vis:set[State] = set()
        vis.add(self.s)
        while pq:
            node = heappop(pq)
            if node.solvability==False:
                return -1
            if node.isGoal:
                return node.gx
            neighbor:list[State] = node.expand()
            for ngh in neighbor:
                if ngh not in vis:
                    vis.add(ngh)
                    heappush(pq,ngh)
        return -1



```
